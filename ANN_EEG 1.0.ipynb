```python
from sklearn.model_selection import train_test_split,cross_val_score
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data=pd.read_csv("Sayantan Combine Data Set.csv")
data.head(5)
data.dropna(inplace=True)
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
from sklearn.metrics import accuracy_score,f1_score,precision_score,confusion_matrix

scaler=StandardScaler()
encoder=OneHotEncoder(sparse_output=False)
```

```python
data.head(5)
```

```python
data['Class'].value_counts()
```

```python
x=data.iloc[:,:20].values
```

```python
y=data.iloc[:,20:].values.reshape(-1,1)
```

```python
y
```

```python
y=encoder.fit_transform(y)
```

```python
results_dict={
    0:"Rage",
    1:"Satisfaction",
    2:"Urge"
}
```

```python
x
```

```python
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
y_train
```

```python
ann=Sequential()
ann.add(Dense(units=10,activation='relu',input_shape=(20,)))
ann.add(Dense(units=10,activation='relu'))
ann.add(Dense(units=3,activation='softmax'))

ann.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
ann.fit(x_train,y_train,epochs=100,batch_size=100)
```

```python
loss, accuracy = ann.evaluate(x_test, y_test)
```
