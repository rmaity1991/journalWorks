```python
from sklearn.model_selection import KFold,StratifiedKFold,train_test_split,GridSearchCV

X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.3,random_state=20)
```

```python
X_Train
```

```python
cv=KFold(n_splits=10)
from sklearn.linear_model import LogisticRegression
logistic_model=LogisticRegression()

logistic_model.fit(X_Train,Y_Train)
```

```python
Y_Pred=logistic_model.predict(X_Test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,RocCurveDisplay,roc_auc_score,precision_recall_curve,recall_score,precision_score,f1_score

print(accuracy_score(Y_Test,Y_Pred))
print(recall_score(Y_Test,Y_Pred,average="macro"))
print(precision_score(Y_Test,Y_Pred,average="macro"))
print(f1_score(Y_Test,Y_Pred,average="macro"))
print(roc_auc_score(Y_Test,logistic_model.predict_proba(X_Test), multi_class='ovr'))
print(confusion_matrix(Y_Test,Y_Pred))
```

```python
logistic_model=LogisticRegression(multi_class="multinomial")
logistic_model.fit(X_Train,Y_Train)
```

```python
Y_Pred=logistic_model.predict(X_Test)
accuracy_score(Y_Test,Y_Pred)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(logistic_model,X_Train,Y_Train,cv=cv)
```
