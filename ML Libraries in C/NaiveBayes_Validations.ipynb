```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler
np.set_printoptions(threshold=np.inf)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
data=pd.read_csv("./eeg data emotion/DigitalAd_dataset.csv")
```

```python
X_Class1=data[data['Status']==0]
X_Class2=data[data['Status']==1]

scaler=MinMaxScaler()
```

```python
X_Class1=X_Class1.sample(143)
```

```python
X_Class2.shape
```

```python
X_Class=pd.concat((X_Class1,X_Class2))
```

```python
X_Class.shape
```

```python
X=X_Class.iloc[:,:2].values
Y=X_Class.iloc[:,2:].values
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.naive_bayes import GaussianNB

model=GaussianNB()
model.fit(X,Y)

Y_pred=model.predict(X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y,Y_pred)
```
