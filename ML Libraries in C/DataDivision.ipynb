```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder,MinMaxScaler
np.set_printoptions(threshold=np.inf)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
data_anger=pd.read_csv("./eeg data emotion/anger.csv").sample(50,random_state=0)
data_fear=pd.read_csv("./eeg data emotion/fear.csv").sample(50,random_state=0)
data_happy=pd.read_csv("./eeg data emotion/happy.csv").sample(50,random_state=0)
data_sad=pd.read_csv("./eeg data emotion/sad.csv").sample(50,random_state=0)
```

```python
data_anger['Class']=0
data_fear['Class']=1
data_happy['Class']=2
data_sad['Class']=3
```

```python
X_Anger=data_anger.iloc[:,:19].values
X_Fear=data_fear.iloc[:,:19].values
X_Happy=data_happy.iloc[:,:19].values
X_Sad=data_sad.iloc[:,:19].values

minmax=MinMaxScaler()

X_Anger=minmax.fit_transform(X_Anger)
X_Fear=minmax.fit_transform(X_Fear)
X_Happy=minmax.fit_transform(X_Happy)
X_Sad=minmax.fit_transform(X_Sad)
```
